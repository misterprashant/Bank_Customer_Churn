```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/churn.csv")

# Show first rows
df.head()
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="Exited", data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
sns.countplot(
    x="Geography",
    hue="Exited",
    data=df
)

plt.title("Geography vs Churn")

plt.show()
```

```python
sns.countplot(
    x="Gender",
    hue="Exited",
    data=df
)

plt.title("Gender vs Churn")

plt.show()
```

```python
sns.boxplot(
    x="Exited",
    y="Age",
    data=df
)

plt.title("Age vs Churn")

plt.show()
```

```python
sns.boxplot(
    x="Exited",
    y="Balance",
    data=df
)

plt.title("Balance vs Churn")

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```
